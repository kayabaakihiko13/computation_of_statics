```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
```

```python
df=pd.DataFrame({
    'tinggi_badan':[168,150,172],
    'berat_badan':[50,38,75],})
df
```

```python
# masukan kostan
constant_variabel=sm.add_constant(df.tinggi_badan)
constant_variabel
```

```python
help(sm.OLS)
```

```python
modeling=sm.OLS(df.berat_badan,df.tinggi_badan,data=df).fit()
```

```python
modeling.summary()
```
