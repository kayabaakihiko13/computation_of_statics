## Bootraping / resample
> boostraping bisa disebut dengan resample dimana resample tersebut untuk mengukur berapa   akurat pada estimasi data

> teknik tersebut digunakan untuk mengukur distribusi sample tersebut

```python
# Import Pakage
import seaborn as sns
import numpy as np
from model.nerve import *
import matplotlib.pyplot as plt


# import plotly.express as px
%matplotlib inline
```

```python
sample = ([7, 11, 8, 30])
```

```python
resample_mean,mean_resample,var_mean_resample,mse_mean_resample,ci_mean_resample=boostraping(sample,1000,70,np.mean).model_resample()
print('Real Mean :',np.mean(sample))
print('Mean Re-Sample:{:.3f}'.format(mean_resample))
print('MSE Re-sample:{:.3f}'.format(mse_mean_resample))
print('CI Re-sample:{}'.format(ci_mean_resample))

```
