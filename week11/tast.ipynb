```python
# bernouli probalilty
import random as r
def pmf(x,p):
    f=p**r.choice(x)*(1-p)**(1-x)
    return f
def mean(p):
    return p
def var(p):
    return p*(1-p)
def std(p):
    return var(p)**(1/2)
def rvs(p,size=1):
    rvs = np.array([])
    for _ in range(0,size):
        if np.random.rand() <= p:
            a=1
            rvs = np.append(rvs,a)
        else:
            a=0
            rvs = np.append(rvs,a)
    return rvs
        
```

```python
own=rvs(.2,100)
```
