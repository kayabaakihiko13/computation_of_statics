```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.random.uniform(size=1000)
def inverse(data,lambda_):
    return -np.log(data)/lambda_
```

```python
# in dari kita
sns.distplot(inverse(data,1))

# hasil dari default numpy
sns.distplot(np.random.exponential(size=1000))

```
